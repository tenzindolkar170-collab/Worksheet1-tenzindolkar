```python
import pandas as pd
```

Sorting

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 19, 22, 25],
    'Score': [88, 92, 85, 95],
}

df = pd.DataFrame(data)
print(df)
```

```python
df_sorted_by_age = df.sort_values(by = 'Age')

print(df_sorted_by_age.head())
```

```python
sorted_by_index = df.sort_index()
print(sorted_by_index.head())
```

```python
subset_iloc = df.iloc[1:3, 0:2]
print("Subset using iloc:")
print(subset_iloc.head())
```

```python
first_two_rows = df.iloc[1:2]
print("\nFirst two rows:")
print(first_two_rows.head())
```

```python
subset_loc = df.loc[df['Age'] > 20, ['Name', 'Score']]
print("\nSubset using loc:")
print(subset_loc)
```

```python
subset_brackets = df[['Name', 'Age']]
print("\nSubset using brackets:")
print(subset_brackets)
```

```python
name_column = df["Name"]
print("\nName column:")
print(name_column)
```

```python
filter_single = df[df["Age"] > 20]
print("\nFilter single column:")
print(filter_single)
```

```python
Filtered_multiple = df[(df["Age"] > 20) & (df["Score"] > 90)]
print("\nFiltered multiple columns:")
print(Filtered_multiple)
```

Another Data Frame

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'HR'],
    'Score': [88, 92, 85, 95]
}

df = pd.DataFrame(data)
print(df)
```

```python
filtered_single = df[df["Department"] == "HR"]
print("\nFiltered single column:")
print(filtered_single)
```

```python
filtered_multiple = df[df['Department'].isin(['HR', 'IT'])]
print(filtered_multiple)
```
